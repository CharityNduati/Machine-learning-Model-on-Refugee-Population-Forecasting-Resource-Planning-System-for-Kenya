2. DATA CLEANING

```python
#filter for assylum in kenya
df = df[df['asylum_location_code']=='KEN']
df
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
#drop 'all' on gender
df = df[df['gender'] != 'all']
print(df['gender'].unique())
```

```python
#remove missing values from min_age & max_age column
df=df.dropna(subset=['min_age','max_age'])
df.head(20)
```

```python
df.shape
```

```python
#remove 'all' in age range
df = df[df['age_range'] != 'all']
print(df['age_range'].unique())
```

```python
df['age_range'].unique()
```
